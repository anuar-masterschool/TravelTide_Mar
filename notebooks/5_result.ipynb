# Joining of groups in single csv file

```python
import numpy as np
import pandas as pd
import numpy as np
```

```python
user_features = pd.read_csv('../data/processed/user_features.csv')
```

```python
user_features
```

```python
user_features['group'] = None
```

```python
user_features
```

```python
family_travelers = pd.read_csv('../data/groups/family_traveler.csv')
user_features.loc[user_features["user_id"].isin(family_travelers['user_id']), 'group'] = 'family_travelers'
```

```python
low_income = pd.read_csv('../data/groups/low_income.csv')
user_features.loc[user_features["user_id"].isin(low_income['user_id']), 'group'] = 'low_income'
```

```python
luxus = pd.read_csv('../data/groups/luxus.csv')
user_features.loc[user_features["user_id"].isin(luxus['user_id']), 'group'] = 'luxus'
```

```python
no_booker = pd.read_csv('../data/groups/no_bookers_group.csv')
user_features.loc[user_features["user_id"].isin(no_booker['user_id']), 'group'] = 'no_booker'
```

```python
overcarriers = pd.read_csv('../data/groups/overcarriers.csv')
user_features.loc[user_features["user_id"].isin(overcarriers['user_id']), 'group'] = 'overcarriers'
```

```python
seniors = pd.read_csv('../data/groups/seniors.csv')
user_features.loc[user_features["user_id"].isin(seniors['user_id']), 'group'] = 'seniors'
```

```python
singles = pd.read_csv('../data/groups/singles.csv')
user_features.loc[user_features["user_id"].isin(singles['user_id']), 'group'] = 'singles'
```

```python
young = pd.read_csv('../data/groups/young.csv')
user_features.loc[user_features["user_id"].isin(young['user_id']), 'group'] = 'young'
```

```python
user_features.to_csv("results/users_with_groups.csv", index=False)
```
